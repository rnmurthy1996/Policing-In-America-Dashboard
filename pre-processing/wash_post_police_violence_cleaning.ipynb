```
# Load data into data
shootings_wash_post = pd.read_csv(file_path)
print("Number of null cities: ", shootings_wash_post["city"].isnull().sum())
shootings_wash_post.head()
```

```
for col in shootings_wash_post.columns:  # iterate over all columns
  if shootings_wash_post[col].dtype not in [np.int64, np.float64]:  # ignore integer and float columns
    # print a header, all unique values, and a separator
    print('Unique Values for {}'.format(col))  
    print(shootings_wash_post[col].unique())   
    print('')
```

```
# Create a year column which will be derived from the date entry
shootings_wash_post.insert(1, "Year", shootings_wash_post["date"])
shootings_wash_post.head()
```

```
# Update the new year column to just contain the 4-digit year of the date
shootings_wash_post = pd.read_csv(file_path)
if not ("Year" in shootings_wash_post):
  shootings_wash_post.insert(0, "Year", shootings_wash_post["date"])
  shootings_wash_post.head()
  for i in range(shootings_wash_post["Year"].size):
    date = shootings_wash_post["Year"][i]
    shootings_wash_post["Year"][i] = int(date[date.index("/", -5)+1:])

shootings_wash_post.head()
```

```
# Lits races
r = shootings_wash_post["race"].unique()
r
```

```
# Drop rows where City is null, since City is part of primary key
shootings_wash_post = shootings_wash_post[shootings_wash_post["city"].notnull()]
print("Number of null cities: ", shootings_wash_post["city"].isnull().sum())

# NOTE: No cities are null, so this should have no effect on the dataset

# Drop rows where Victim name is "TK TK", which appears to be what is put when the victim name is not known
shootings_wash_post = shootings_wash_post[shootings_wash_post["name"] != "TK TK"]

# Drop any remaining duplicates based on key
shootings_wash_post = shootings_wash_post.drop_duplicates(subset=['Year', 'city', 'name'], keep='last')
```

```
# Add a "Killed by gun wound column"

# NOTE in this dataset all killings are the result of being shot
if not ("is_gun_death" in shootings_wash_post):
  shootings_wash_post.insert(1, "is_gun_death", True)

shootings_wash_post = shootings_wash_post.replace({'is_gun_death': {True: 'y', False: 'n'}})
# print("Number of gunshot deaths: ", shootings_wash_post["is_gun_death"].sum() )
```

```
# Check whether the columns "Year", "City", "Victim's name" can be used as Primary Keys

# Step 1: Group dataframe by candidate columns 
grouped = shootings_wash_post.groupby(["Year", "city", "name"]) 
# Step 2: Count rows in each group
counts = grouped.size()

# Note, since we are using Victim's name in the primary key, we should remove all cases where name is "Name withheld by police"
print(counts[counts > 1])

```

```
# Add a county column using a US States database

# Set file path
states_file_path = '/content/drive/MyDrive/CIS 550/Project/uscities.csv'
cities = pd.read_csv(states_file_path)
cities.head()

shootings_wash_post_with_county = shootings_wash_post.merge(cities, left_on=["city", "state"], right_on=["city", "state_id"], how="left")
shootings_wash_post_with_county.head()

print("Number of null counties: ", shootings_wash_post_with_county["county_name"].isnull().sum())

# Drop rows where County is null
shootings_wash_post_with_county = shootings_wash_post_with_county[shootings_wash_post_with_county["county_name"].notnull()]
shootings_wash_post_with_county.head()
```

```
shootings_wash_post_with_county.head()

# Drop unwanted columns
shootings_wash_post_with_county_reduced_columns = shootings_wash_post_with_county.drop(['armed', 'signs_of_mental_illness', 'threat_level', 'flee', 'body_camera', 'city_ascii', 'state_id', 'state_name', 'county_fips', 'lat', 'lng', 'population', 'density', 'source', 'military', 'incorporated', 'timezone', 'ranking', 'zips', 'id_y', 'id_x'], axis=1)
shootings_wash_post_with_county_reduced_columns.head()
```

```
# Normalize column headings
shootings_wash_post_with_county_reduced_columns_renamed = shootings_wash_post_with_county_reduced_columns.rename(columns={"Year": "year", "gender": "sex", "county_name": "county", "manner_of_death": "cause_of_death"})

shootings_wash_post_with_county_reduced_columns_renamed.insert(9, "article", "NULL")
shootings_wash_post_with_county_reduced_columns_renamed = shootings_wash_post_with_county_reduced_columns_renamed[['year', 'city', 'name', 'age', 'sex', 'race', 'cause_of_death', 'date', 'county', 'state', 'is_gun_death', 'article']]
shootings_wash_post_with_county_reduced_columns_renamed.head()
```

```
# Normalize Race Names
race_map = {
    'W': 'White',
    'B': 'Black',
    'A': 'Asian',
    'N': 'Native American',
    'H': 'Hispanic',
    'nan': 'Unknown',
    'O': 'Unknown',
}

shootings_wash_post_with_county_reduced_columns_renamed_1 = shootings_wash_post_with_county_reduced_columns_renamed.replace({"race": race_map})
shootings_wash_post_with_county_reduced_columns_renamed_1.head()
r = shootings_wash_post_with_county_reduced_columns_renamed_1["race"].unique()
r
```
